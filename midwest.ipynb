```python
import pandas as pd
import numpy as np
```

```python
mw = pd.read_csv('midwest.csv')
mw
```

```python
mwc = mw.copy()
mwc
```

```python
mwc = mwc.rename(columns = {'poptotal':'total', 'popasian':'asian'})
mwc
```

```python
mwc['asianper'] = (mwc['asian'] / mwc['total']) * 100
mwc
```

```python
mwc['asianper'].plot.hist()
```

```python
mwc.describe()
```

```python
mwc['test'] = np.where(mwc['asianper'] >= 0.487246, 'large', 'small')
mwc
```

```python
count_test = mwc['test'].value_counts()
```

```python
count_test
```

```python
count_test.plot.bar(rot=0)
```
